4. Chuẩn hóa dữ liệu theo min max

```python
df3 = df1.copy()
# sepal.length	sepal.width	petal.length	petal.width
sepal_length_min = df3['sepal.length'].min()
sepal_length_max = df3['sepal.length'].max()

sepal_width_min = df3['sepal.width'].min()
sepal_width_max = df3['sepal.width'].max()

petal_length_min = df3['petal.length'].min()
petal_length_max = df3['petal.length'].max()

petal_width_min = df3['petal.width'].min()
petal_width_max = df3['petal.width'].max()
```

```python
df3['sepal.length'] = (df3['sepal.length'] - sepal_length_min)/(sepal_length_max-sepal_length_min)
df3['sepal.width'] = (df3['sepal.width'] - sepal_width_min)/(sepal_width_max-sepal_width_min)
df3['petal.length'] = (df3['petal.length'] - petal_length_min)/(petal_length_max-petal_length_min)
df3['petal.width'] = (df3['petal.width'] - petal_width_min)/(petal_width_max-petal_width_min)
```

6. Xây dựng mô hình hồi quy tuyến tính để đoán kết quả của thuộc tính 
variety  dựa trên các thuộc tính khác

```python
df5 = df3.copy()
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

X = df5.drop("variety", axis=1).values
y = df5["variety"].values
y = np.resize(y, (150,1))

enc = OneHotEncoder(handle_unknown='ignore').fit(y)
y_one_hot = enc.transform(y).toarray()
# y_one_hot.shape
y_one_hot
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_one_hot, test_size=0.2)
```

```python
lireg = LinearRegression()
lireg.fit(X_train,y_train)
```

```python
y_predict1 = lireg.predict(X_test)
```

```python
y_predict = y_predict1
for i in range(len(y_test)):
    yy = max(y_predict[i])
    for j in range(len(y_test[i])):
        if yy == y_predict[i][j] :
            y_predict[i][j]=1
        else:
            y_predict[i][j]=0
print(y_predict)
```

```python
y_pred = enc.inverse_transform(y_predict)
```

```python
yy_pre = np.resize(y_pred, 150)
yy_pre
```
