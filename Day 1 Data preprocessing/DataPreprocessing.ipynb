## Câu 4: Chuẩn hóa dữ liệu theo min max
<img src="Min_Max_Normalization.png">

```python
df3 = df1.copy()
df3
```

```python
age_min = df3.Age.min()
age_max = df3.Age.max()
income_min = df3.Income.min()
income_max = df3.Income.max()
```

```python
df3.Age = (df3.Age - age_min)/(age_max - age_min)
df3.Income = (df3.Income - income_min)/(income_max - income_min)
```

```python
df3
```

## Câu 6: Xây dựng mô hình hồi quy tuyến tính để đoán kết quả của thuộc tính Online Shopper dựa trên các thuộc tính khác

```python
df5 = df4.copy()
```

```python
df5['Online Shopper'].replace({"No":0,"Yes":1},inplace=True)
df5
```

```python
X = df5.drop("Online Shopper", axis=1)
y = df5["Online Shopper"]
thread_point = int(0.8 * len(df5))
```

```python
# Giải thích threatpoint, dấu : ở trước và sau threatpoint có ý nghĩa
# [1, 2, 3, 4, 5, 6, 7]
# {0, 1, 2, 3, 4, 5, 6}
# thread_point = int(0.8 * len(df5) )
print(thread_point)
```

```python
# X[0, 1, 2, 3, 4, 5, 6, 7, 8] -> X[0:8] -> X[:8]
# X[9] -> X[8:9] -> X[8:]
```

```python
X_train, X_test = X[:thread_point], X[thread_point:]
y_train, y_test = y[:thread_point], y[thread_point:]
```

```python
from sklearn.linear_model import LogisticRegression
lireg = LogisticRegression()
lireg.fit(X_train,y_train)
```

```python
print("Predicted label: {}, True label: {}".format(lireg.predict(X_test), y_test.values))
```

```python
lireg.score(X_test,y_test)
```
